```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
import tensorflow as tf
from mlxtend.plotting import plot_decision_regions
```

```python
np.random.seed(0)
```

```python
n_pts = 500
X, y = datasets.make_circles(n_samples=n_pts, noise=0.1, factor=0.2)
```

```python
plt.scatter(X[y==0, 0],X[y==0, 1])
plt.scatter(X[y==1, 0],X[y==1, 1])
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(4, input_shape=(2,), activation='sigmoid'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
model.compile(tf.keras.optimizers.Adam(lr=0.01), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
h = model.fit(x=X, y=y, verbose=1, batch_size=20, epochs = 70, shuffle=True)
```

```python
plt.plot(h.history['accuracy'])
plt.xlabel('epoch')
plt.legend(['accuracy'])
plt.title('accuracy')
```

```python
plt.plot(h.history['loss'])
plt.xlabel('epoch')
plt.legend(['loss'])
plt.title('loss')
```

```python
plot_decision_regions(X,y, clf=model, legend=2);
```
